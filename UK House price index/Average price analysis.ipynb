```python
# Import libraries 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, YearLocator
%matplotlib inline
```

```python
url_UKHousePriceIndex = 'https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xlsx'

average_price = pd.read_excel(url_UKHousePriceIndex, sheet_name='Average price', index_col= None)
```

```python
average_price.head()
```

```python
#Transpose the columns and rows 
average_price = average_price.transpose()
average_price.columns = average_price.iloc[0]
average_price = average_price.iloc[1:]
average_price = average_price.reset_index()
average_price.head()
```

```python
#Renaming the columns
average_price.rename(columns={'index': 'Borough', average_price.columns[1]: 'Entry'}, inplace=True)
average_price.columns.name = '' #I didnt know what Unnamed : 0 meant so I removed it
average_price.head()
```

```python
#Melting the Dataframe
average_price = pd.melt(average_price, id_vars=['Entry', 'Borough'], var_name='Year/Month', value_name='Housing Price')

#Ensuring data types are correct by converting average prices to floating point numbers
average_price['Housing Price'] = pd.to_numeric(average_price['Housing Price'])
average_price.dtypes
average_price.head()

```

```python
#Inspect the Borough column and drop any null values 
average_price.dropna(inplace=True)
```

```python
plt.figure(figsize=(12,5))
for region,group in average_price.groupby(['Borough']):
    plt.plot(group['Year/Month'],
             group['Housing Price'],
             label=region)
plt.legend()
plt.title('Average Price of Properties in London Boroughs')
plt.ylabel('Average Price (£)')
plt.xlabel('Transaction Date')
plt.xticks(rotation=-90)
plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(YearLocator())
plt.show()
```
